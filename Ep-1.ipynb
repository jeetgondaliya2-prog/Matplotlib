```python
import numpy as np                             # matplpotlib basics
import pandas as pd  
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5,11)
x
```

```python
y=x**2
y
```

```python
plt.plot(x,y)                                # plotting of graph
plt.title('Hello World')                     # title of graph
plt.xlabel('x-axis')                         # x axis label
plt.ylabel('y-axis')                         # y axis label
```

```python
plt.subplot(2,2,1)       # subplot (x,y,z) where x,y indicate the rows and cols and z indicate choosing of that plot
plt.plot(x,x)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
```

```python
fig=plt.figure()                                              # object oriented way
axis1=fig.add_axes([0.1,0.1,0.4,0.8])                         # 40% width and 80% height
axis1.plot(x,y)
axis2=axis1=fig.add_axes([0.2,0.4,0.2,0.2])                   # 20% width and 20% height , 20% from y-axis and 40% from x axis
axis2.plot(x,y)
```

```python
plt.scatter(x,y)                                            # scatter plot
```

```python
plt.hist(x)                                 # histogram
```

```python
plt.boxplot(x)                                  # boxplot
```
