```python
import numpy as np                              # optional designing
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5,11)
```

```python
x
```

```python
y=x**2
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,color='red')             # change color of line
```

```python
plt.plot(x,y,color='red',marker='+')            # add marking points
```

```python
plt.plot(x,y,color='red',marker='o',linestyle='--')           # changing linestyle
```

```python
plt.plot(x,y,color='red',marker='o',linestyle='--',markerfacecolor='green')   # change color of marking
```
